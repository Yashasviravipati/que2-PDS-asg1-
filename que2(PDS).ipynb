```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('StudentsPerformance.csv')

# Display first few rows to check
data.head()
```

**Create Visualizations**

**1. Histogram:** **Distribution of Math Scores bold text**

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['math score'], bins=20, kde=True, color='blue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

**2. Boxplot: Gender-wise Performance in Reading Scores**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='gender', y='reading score', data=data, palette='coolwarm')
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```

**3. Heatmap: Correlation between Different Scores**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Scores')
plt.show()
```

**4. Boxplot: Parental Education vs. Math Scores**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='math score', data=data, palette='viridis')
plt.xticks(rotation=45)
plt.title('Math Scores by Parental Education')
plt.xlabel('Parental Education Level')
plt.ylabel('Math Score')
plt.show()
```

**5. Violin Plot: Lunch Type vs. Writing Score**

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='lunch', y='writing score', data=data, palette='Set2')
plt.title('Writing Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Writing Score')
plt.show()
```
